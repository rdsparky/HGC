```python
import tensorflow as tf
import os
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
model = tf.keras.models.load_model("48_CNN.model")
```

```python
CATEGORIES = ["female", "male"]
```

```python
def prepare(filepath):
    IMG_SIZE = 48  # 50 in txt-based
    img_array = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE),)
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)
```

```python
prediction = model.predict([prepare("C:/Users/Spark/Desktop/t6.jpg")])
ta = cv2.imread("C:/Users/Spark/Desktop/t6.jpg" ,cv2.IMREAD_GRAYSCALE )
plt.imshow((ta))  # graph it
plt.show()  # display!
print(prediction)  # will be a list in a list.
print(CATEGORIES[int(prediction[0][0])])
```
